# Variables númericas

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_palette("husl", df.shape[1])

df.hist(bins=50, figsize=(20, 15))
plt.show()
```

# Valores Categóricas


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear una paleta personalizada en la gama de rosa y morado
paleta_rosa_morado = sns.color_palette(["#FF69B4", "#DB7093", "#C71585", "#800080", "#DA70D6", "#BA55D3", "#EE82EE"])

# Gráfico de conteo para autores
plt.figure(figsize=(10, 6))
sns.countplot(y=df['Authors'], order=df['Authors'].value_counts().index[:10], palette=paleta_rosa_morado)
plt.title('Top 10 autores Más Frecuentes')
plt.xlabel('Frecuencia')
plt.ylabel('Libros')
plt.show()

# Gráfico de conteo para publicistas
plt.figure(figsize=(10, 6))
sns.countplot(y=df['Publisher'], order=df['Publisher'].value_counts().index[:10], palette=paleta_rosa_morado)
plt.title('Top 10 Publicistas Más Frecuentes')
plt.xlabel('Frecuencia')
plt.ylabel('Publicistas')
plt.show()
```
